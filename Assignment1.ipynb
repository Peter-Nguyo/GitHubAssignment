```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Loading dataset
file_path = "KRA_Class_Dataset.xlsx"
df = pd.read_excel(file_path)
```

```python
#Bar chart
plt.figure(figsize=(10, 6))
df.groupby('sector')['tax_paid'].sum().sort_values().plot(kind='bar', color='skyblue')
plt.title('Total Tax Paid per Sector')
plt.xlabel('Sector')
plt.ylabel('Total Tax Paid')
plt.xticks(rotation=45)
plt.show()
```

```python
#Pie chart
plt.figure(figsize=(6, 6))
df['taxpayer_type'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['lightcoral','grey', 'lightblue'])
plt.title('Proportion of Taxpayer Types')
plt.ylabel('')  # Remove y-axis label
plt.show()
```

```python
#Histogram
plt.figure(figsize=(8, 6))
plt.hist(df['tax_paid'], bins=20, edgecolor='black', color='blue')
plt.title('Distribution of Tax Paid')
plt.xlabel('Tax Paid')
plt.ylabel('Frequency')
plt.show()
```
